```python
#importing Libraries
import numpy as np#for data analysis
import pandas as pd#for data visualization
import selenium
import time
from bs4 import BeautifulSoup
#Import selenium webdriver
from selenium import webdriver
#importing required Exceptions which needed to be handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
#importing requests
import requests
#importing regex
import re
```

```python
# Activating the chrome browser
driver = webdriver.Chrome(r'E:/chromedriver')
```

```python
# Opening the homepage of olx.in
driver.get('https://www.olx.in/cars_c84')
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')
```

```python
brand=[]
variant=[]
fueltype=[]
mileage=[]
model=[]
year=[]
owner=[]
transmission=[]
price=[]
```

```python
#scraping the price
prices=driver.find_elements_by_xpath("//span[@class='_89yzn']")
for i in prices:
    if i.text is None :
        price.append("--") 
    else:
        price.append(i.text)
```

```python
urls=driver.find_elements_by_xpath("//a[@class='fhlkh']")
UR=[]
for i in urls:
    UR.append(i.get_attribute('href'))
for url in UR:
    driver.get(url)
    try:
        brands=driver.find_element_by_xpath("//span[@data-aut-id='value_make']")
        brand.append(brands.text)
        models=driver.find_element_by_xpath("//span[@data-aut-id='value_model']")
        model.append(models.text)
        years=driver.find_element_by_xpath("//span[@data-aut-id='value_year']")
        year.append(years.text)
        fuels=driver.find_element_by_xpath("//span[@data-aut-id='value_petrol']")
        fueltype.append(fuels.text)
        variants=driver.find_element_by_xpath("//span[@data-aut-id='value_variant']")
        variant.append(variants.text)
        transmissions=driver.find_element_by_xpath("//span[@data-aut-id='value_transmission']")
        transmission.append(transmissions.text)
        mileages=driver.find_element_by_xpath("//span[@data-aut-id='value_mileage']")
        mileage.append(mileages.text)
        owners=driver.find_element_by_xpath("//span[@data-aut-id='value_owner']")
        owner.append(owners.text)
       
    except NoSuchElementException   as e:
        brand.append("NO rating")
        variant.append("NO rating")
        fueltype.append("NO rating")
        mileage.append("NO rating")
        model.append("No rating")
        transmission.append("No rating")
        year.append("No rating")
        owner.append("NO rating")
        
```
